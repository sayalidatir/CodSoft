# Simple To-DO List:

```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk


def add_task():
    task = e1.get()
    if task:
        listbox.insert(tk.END, task)
        e1.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Please enter a task.")


def remove_task():
    index = listbox.curselection()
    if index:
        listbox.delete(index)
    else:
        messagebox.showwarning("Warning", "Please select a task to remove.")


root = tk.Tk()
root.title("My Simple To-Do List")
root.minsize(width=400, height=550)
root.maxsize(width=400, height=400)

root.configure(bg='white')


image = Image.open("todolist.PNG")
image = image.resize((200, 200))
tk_image = ImageTk.PhotoImage(image)
image_label = tk.Label(root, image=tk_image)
image_label.pack()


e1 = tk.Entry(root, font=('calibri', 13), bd=3, width=30)
e1.pack(pady=10)

add_button = tk.Button(root, text="Add Task",bg='#00acee',fg='white',bd=4, command=add_task)
add_button.place(x=100,y=255)
remove_button = tk.Button(root, text="Remove Task",bg='#FF0000',fg='white',bd=4, command=remove_task)
remove_button.place(x=200,y=255)



listbox = tk.Listbox(root, selectmode=tk.SINGLE, width=35,font=('calibri', 13), bd=3)
listbox.pack(pady=40)


root.mainloop()
```
